```python
from sklearn.metrics import accuracy_score, r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.width', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

# neu la chuoi nghich dao
def isPalindrome(s):
    return s == s[::-1]

# df_train = pd.read_csv('test.csv')
# doc du lieu
df_train = pd.read_excel('train_data.xlsx')
df_test = pd.read_excel('test_data.xlsx')
```

```python
sim_array = df_train['sim_number'].astype(str).values.tolist()
chu_so = []
dem_so_sau_tam = []
dem_so_chin = []
nghich_dao = [] #3
phat_loc = [] #4
tien_len_chan = [] #5
sinh_tai_loc_phat = [] #6
phat_tai_phat_loc = [] #7
than_tai_nho = [] #8
than_tai_lon = [] #9
ong_dia_nho = [] #10
ong_dia_lon = [] #11
tam_hoa = [] #12
luc_quy_giua = [] #13
luc_quy_duoi = [] #14
ngu_quy_giua = [] #15
ngu_quy_duoi = [] #16
tu_quy_giua = [] #17
tu_quy_duoi = [] #18
san_bang_tat_ca = [] #19
bon_mua_khong_that = [] #20
doc_nhat_vo_nhi = [] #21
soi_guong = [] #22
soi_guong_vip = [] #23
sanh_lui = []
duoi_sanh_lui = []
sanh_tien = [] #24
duoi_sanh_tien = [] #25
lap_doi = [] #26
lap_tam = [] #27
lap_tu = [] #28
nhi_tam_duoi = [] #29
nhi_tam_giua = [] #30
nhi_tu_duoi = [] # 31
nhi_tu_giua = [] #32
nha_mang = [] #33
tien_kep_duoi = [] #34
tien_kep_giua = [] #35
duoi_tien_kep = [] #36

# ma regex bo nhi tam, vi du: 131313
regexNhiTam = re.compile(r"(\d)\1{1}(\d)\2{1}(\d)\3{1}")
# ma regex bo nhi tu, vi du: 14141414
regexNhiTu = re.compile(r"(\d)\1{1}(\d)\2{1}(\d)\3{1}(\d)\4{1}")
# ma regex luc quy, vi du: 666666
regexLuc = re.compile(r"(\d)\1{5}")
# ma regex ngu quy, vi du: 55555
regexNgu = re.compile(r"(\d)\1{4}")
# ma regex tu quy, vi du: 4444
regexTu = re.compile(r"(\d)\1{3}")

viettel = ['96','97','98','86','32','33','34','35','36','37','38','39']#1
vinaphone = ['88','91','94','83','84','85','81','82']#2
mobifone = ['89','90','93','70','79','77','76','78']#3
vietnamobile = ['92','56','58', '52']#4
gmobile = ['99','59']#5
itel = ['87']#6

for idx in range(len(sim_array)):
    # tim nha mang
    if (sim_array[idx][:2] in viettel):
        nha_mang.append('viettel')
    elif (sim_array[idx][:2] in vinaphone):
        nha_mang.append('vinaphone')
    elif (sim_array[idx][:2] in mobifone):
        nha_mang.append('mobifone')
    elif (sim_array[idx][:2] in vietnamobile):
        nha_mang.append('vietnamobile')
    elif (sim_array[idx][:2] in gmobile):
        nha_mang.append('gmobile')
    elif (sim_array[idx][:2] in itel):
        nha_mang.append('itel')

    # dem so chu so
    chu_so.append(1 if len(set(sim_array[idx])) < 5 else 0)

    # dem so sau va tam
    dem_so_sau_tam.append(len(re.findall('6', sim_array[idx])) + len(re.findall('8', sim_array[idx])))

    # dem so chin
    dem_so_chin.append(len(re.findall('9', sim_array[idx])))

    # kiem tra sim nghich dao
    nghich_dao.append(1 if isPalindrome(sim_array[idx]) else 0)

    # kiem tra sim phat loc
    if sim_array[idx][-2:] == '68' or sim_array[idx][-2:] == '86':
        # dem sim phat loc
        pl = 0
        # dem sim tien len chan
        tlc = 0
        # dem sim sinh tai loc phat
        stlp = 0
        # findall la tim so lan xuat hien cua chuoi can tim trong chuoi cha, vi du o duoi tim so lan xuat hien của chuoi '68' trong sim_aray[idx]
        pl += len(re.findall('68', sim_array[idx]))
        pl += len(re.findall('86', sim_array[idx]))
        if sim_array[idx][-4:-2] == '24':
            tlc += len(re.findall('2468', sim_array[idx]))
        if sim_array[idx][-4:-2] == '13':
            stlp += len(re.findall('1368', sim_array[idx]))
        phat_loc.append(pl)
        tien_len_chan.append(tlc)
        sinh_tai_loc_phat.append(stlp)
    else:
        phat_loc.append(0)
        tien_len_chan.append(0)
        sinh_tai_loc_phat.append(0)

    # kiem tra sim phat tai phat loc
    ptpl = len(re.findall('8683', sim_array[idx]))
    phat_tai_phat_loc.append(ptpl)

    # kiem tra sim than tai nho
    if sim_array[idx][-2:] == '39':
        ttn = len(re.findall('39', sim_array[idx]))
        than_tai_nho.append(ttn)
    else:
        than_tai_nho.append(0)

    # kiem tra sim than tai lon
    if sim_array[idx][-2:] == '79':
        ttl = len(re.findall('79', sim_array[idx]))
        than_tai_lon.append(ttl)
    else:
        than_tai_lon.append(0)

    # kiem tra sim ong dia nho
    if sim_array[idx][-2:] == '38':
        odn = len(re.findall('38', sim_array[idx]))
        ong_dia_nho.append(odn)
    else:
        ong_dia_nho.append(0)

    # kiem tra sim ong dia lon
    if sim_array[idx][-2:] == '78':
        odl = len(re.findall('78', sim_array[idx]))
        ong_dia_lon.append(odl)
    else:
        ong_dia_lon.append(0)

    # kiem tra sim tam hoa
    th = 0
    # match la kiem tra chuoi can tim co giong ma regex khong
    if re.compile(r"(\d)\1{2}").match(sim_array[idx][-3:]):
        th = 1
        if re.compile(r"(\d)\1{2}").match(sim_array[idx][-6:-3]) and sim_array[idx][-6:-3] != sim_array[idx][-3:]:
            th = 2
            if re.compile(r"(\d)\1{2}").match(sim_array[idx][:3]) and sim_array[idx][:3] != sim_array[idx][-6:-3]:
                th = 3
        tam_hoa.append(th)
    else:
        tam_hoa.append(0)

    lqd = 0
    nqd = 0
    tqd = 0

    # kiem tra sim luc quy duoi
    if regexLuc.match(sim_array[idx][-6:]):
        lqd = 1
    # kiem tra sim ngu quy duoi
    elif regexNgu.match(sim_array[idx][-5:]):
        nqd = 1
    # kiem tra sim tu quy duoi
    elif regexTu.match(sim_array[idx][-4:]):
        tqd = 1

    luc_quy_duoi.append(lqd)
    ngu_quy_duoi.append(nqd)
    tu_quy_duoi.append(tqd)

    lqg = 0
    nqg = 0
    tqg = 0

    # kiem tra sim luc quy giua
    for i in range(0, 3):
        j = i + 6
        if regexLuc.match(sim_array[idx][i:j]):
            lqg += 1

    # neu khong co luc quy giua, kiem tra ngu quy giua
    if lqg == 0:
        for i in range(0, 4):
            j = i + 5
            if regexNgu.match(sim_array[idx][i:j]):
                nqg += 1

    # neu khong co luc quy giua và ngu quy giua, kiem tra tu quy giua
    if lqg == 0 and nqg == 0:
        for i in range(0, 5):
            j = i + 4
            if regexTu.match(sim_array[idx][i:j]):
                tqg += 1

    luc_quy_giua.append(lqg)
    ngu_quy_giua.append(nqg)
    tu_quy_giua.append(tqg)

    # kiem tra sim san bang tat ca
    san_bang_tat_ca.append(1 if sim_array[idx][-4:] == '6789' else 0)

    # kiem tra sim bon mua khong that
    bon_mua_khong_that.append(1 if sim_array[idx][-4:] == '4078' else 0)

    # kiem tra sim doc nhat vo nhi
    doc_nhat_vo_nhi.append(1 if sim_array[idx][-4:] == '1102' else 0)

    # kiem tra sim soi guong
    if isPalindrome(sim_array[idx][-7:]):
        soi_guong.append(2)
        soi_guong_vip.append(1 if sim_array[idx][-1] == sim_array[idx][-4] else 0)
    elif isPalindrome(sim_array[idx][-6:]):
        soi_guong.append(1)
        soi_guong_vip.append(1 if sim_array[idx][-1] == sim_array[idx][-3] else 0)
    else:
        soi_guong.append(0)
        soi_guong_vip.append(0)

    if int(sim_array[idx][-1]) == int(sim_array[idx][-2]) + 1 and int(sim_array[idx][-2]) == int(sim_array[idx][-3]) + 1:
        i = -1
        count = -2
        while int(sim_array[idx][i]) == int(sim_array[idx][i - 1]) + 1:
            count +=1
            i -= 1
        sanh_tien.append(count)
        duoi_sanh_tien.append(sim_array[idx][-1])
    else:
        sanh_tien.append(0)
        duoi_sanh_tien.append(0)

    if int(sim_array[idx][-1]) == int(sim_array[idx][-2]) - 1 and int(sim_array[idx][-2]) == int(sim_array[idx][-3]) - 1:
        i = 8
        count = -2
        while i > 0 and int(sim_array[idx][i - 1]) > 0 and int(sim_array[idx][i]) == int(sim_array[idx][i - 1]) - 1:
            count +=1
            i -= 1
        sanh_lui.append(count)
        duoi_sanh_lui.append(sim_array[idx][-1])
    else:
        sanh_lui.append(0)
        duoi_sanh_lui.append(0)

    l2 = 0
    l3 = 0
    l4 = 0

    # kiem tra sim lap doi, lap tam, lap tu, chac cung khong quan trong lam
    if th == 0 and tqd == 0 and nqd == 0 and lqd == 0:
        l2 = len(re.findall(sim_array[idx][-2:], sim_array[idx]))
        l3 = len(re.findall(sim_array[idx][-3:], sim_array[idx]))
        l4 = len(re.findall(sim_array[idx][-4:], sim_array[idx]))

    lap_doi.append(l2)
    lap_tam.append(l3)
    lap_tu.append(l4)

    n3d = 0
    n3g = 0
    n4d = 0
    n4g = 0

    if regexNhiTu.match(sim_array[idx][-8:]):
        n4d = 1
    elif regexNhiTam.match(sim_array[idx][-6:]):
        n3d = 1

    if regexNhiTu.match(sim_array[idx][:8]):
        n4g = 1
    elif regexNhiTam.match(sim_array[idx][:6] or sim_array[idx][1:7] or sim_array[idx][2:8]):
        n3g = 1

    nhi_tam_duoi.append(n3d)
    nhi_tam_giua.append(n3g)
    nhi_tu_duoi.append(n4d)
    nhi_tu_giua.append(n4g)

df_train['chu_so'] = chu_so
df_train['dem_so_sau_tam'] = dem_so_sau_tam
df_train['dem_so_chin'] = dem_so_chin
df_train['nghich_dao'] = nghich_dao
df_train['phat_loc'] = phat_loc
df_train['tien_len_chan'] = tien_len_chan
df_train['sinh_tai_loc_phat'] = sinh_tai_loc_phat
df_train['phat_tai_phat_loc'] = phat_tai_phat_loc
df_train['than_tai_nho'] = than_tai_nho
df_train['than_tai_lon'] = than_tai_lon
df_train['ong_dia_nho'] = ong_dia_nho
df_train['ong_dia_lon'] = ong_dia_lon
df_train['tam_hoa'] = tam_hoa
df_train['luc_quy_giua'] = luc_quy_giua
df_train['luc_quy_duoi'] = luc_quy_duoi
df_train['ngu_quy_giua'] = ngu_quy_giua
df_train['ngu_quy_duoi'] = ngu_quy_duoi
df_train['tu_quy_giua'] = tu_quy_giua
df_train['tu_quy_duoi'] = tu_quy_duoi
df_train['san_bang_tat_ca'] = san_bang_tat_ca
df_train['bon_mua_khong_that'] = bon_mua_khong_that
df_train['doc_nhat_vo_nhi'] = doc_nhat_vo_nhi
df_train['soi_guong'] = soi_guong
df_train['soi_guong_vip'] = soi_guong_vip
df_train['sanh_lui'] = sanh_lui
df_train['duoi_sanh_lui'] = duoi_sanh_lui
df_train['sanh_tien'] = sanh_tien
df_train['duoi_sanh_tien'] = duoi_sanh_tien
df_train['lap_doi'] = lap_doi
df_train['lap_tam'] = lap_tam
df_train['lap_tu'] = lap_tu
df_train['nha_mang'] = nha_mang

df_train = pd.get_dummies(df_train, prefix='sim', columns=['nha_mang'])
df_train.to_excel('test_data.xlsx')
```
